```python
import cohere
co = cohere.Client(COHERE_API_KEY)
```

```python
import time
from tqdm import trange

cohere_resp = []

for i in trange(20):
    tmp = {"cohere_resp":None, "index":i, "label":df["label"][i]}
    claim = df["claim"][i]
    response = co.chat(
        message=f"Is the following claim true: {claim}",
        connectors=[{"id": "web-search"}]
    )
    tmp["cohere_resp"] = response.text
    time.sleep(1)
    USER_MESSAGE_2 = "Are you sure?"
    response_2 = co.chat(
        chat_history = [
            {"role": "USER", "message": f"Is the following claim true: {claim}"},
            {"role": "CHATBOT", "message": response.text}
        ],
        message = USER_MESSAGE_2,
        connectors = [{"id": "web-search"}]
    )
    # print(response_2.text)
    tmp["user_message_2"] = USER_MESSAGE_2
    tmp["cohere_resp_2"] = response_2.text
    cohere_resp.append(tmp)
    time.sleep(0.5)
```

```python
import termcolor
df_results = pd.DataFrame(cohere_resp)
for i in range(df_results.shape[0]):
    print("Index: "+str(i), end=".\n")
    print(termcolor.colored("True label:\t","magenta")+ termcolor.colored(df_results["label"][i],"blue"))
    print(termcolor.colored("User:\t\t", "blue")+"Is the following claim true: <claim>")
    print(termcolor.colored("LLM response:\t", "cyan")+df_results["cohere_resp"][i])
    print(termcolor.colored("User:\t\t", "blue")+"Are you sure?")
    print(termcolor.colored("LLM response:\t", "cyan")+df_results["cohere_resp_2"][i])

    print("\n---\n")
```
